## Background: de Broglie's hypothesis

In the beginning of the 20th century, wave-like and particle-like behavior were thought to be incompatible. Objects either acted like a wave or a particle. This notion started to unravel when measurements (see for example the photoelectric effects and ultraviolet catastrophe in black-body radiation) started to suggest that **light is made up of discrete energy packets called photons**.  **The energy of a photon is related to its frequency by Planck‚Äôs constant, $E = h \nu$, and its momentum is inversely proportional to its wavelength $$ p = h/\lambda.$$**  And matter (e.g. electrons, protons, neutrons) was treated as particles following Newton‚Äôs laws, but this was incompatible with the observations that electrons occupied discrete energy levels, which could not be explained using classical mechanics. If classical mechanics holds for electrons, then electrons orbiting around a nucleus should be equivalent to planets orbiting around a star. At small length and energy scales, clearly Newton‚Äôs laws were not sufficient. 
<br><br>
In his 1924 graduate thesis, Louis de Broglie (pronounced "de Broy") proposed that perhaps all matter particles, not just light, behave like waves. If light can sometime behave like a particle, why cannot matter also sometimes behave like a wave? That wavelength is given by the de Broglie relation
$$ \lambda_{\text{de Broglie}} = h/p . $$

The momentum of a particle is given by $p = m v$.  The kinetic energy of a particle is related to its velocity $v$ and mass $m$ by $E = \frac{1}{2}m v^2  = \frac{p^2}{2 m}$.  Solving for momentum in terms of energy, an electron‚Äôs wavelength is related to its energy by  $$\lambda_{\text{de Broglie}} = \frac{h}{\sqrt{2 m E}}.$$   Here $h$ is Planck‚Äôs constant, and ùëö is the mass of the particle. This is the de Broglie relation that relates a particles wavelength to its energy. 


So why can we not just perform the two slit experiment with electrons and try to observe interference? The two slit experiment only works when the slits are separated by a distance $ùëë$ similar to the wavelength. Visible light has a wavelength of approximately 1 ùúáùëö ($10^{-6}$ m). Electrons instead were predicted to have wavelengths on the order of 10 picometers. That is too small of a slit to make.
Fortunately, we can borrow ideas from X-ray scattering of crystals, since X-rays also have wavelengths that are this small. In 1913, W.H. and W.L. Bragg showed that X-rays incident on a crystal are reflected at specific angles. The diffraction of the X-rays by the crystal can be understood by picturing a crystal as containing a periodic set of planes, with each plane reflecting a wave like a simple plane mirror.
