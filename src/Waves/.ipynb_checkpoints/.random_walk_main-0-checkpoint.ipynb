```python
def normal(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-1/2*((x-mu)/sigma)**2)

x = np.linspace(-4,4,100)
y = normal(x, 0, 1)
plt.plot(x,y)
plt.xlabel('Standard Deviations Outside Mean')
plt.ylabel('Probability Density')
plt.show()
```

```python
def random_walk(N, h):
    """
    Simulates a random walk
    :param N: number of steps to take
    :param h: the step size
    """
    steps = np.random.choice([1,-1], N) # discrete steps
    position = np.cumsum(steps)  # absolute position
    position = np.insert(position, 0, 0.0) # insert the initial position
    return position


# generate a random walk
N=400 # number steps
h = 1 # standard deviation of step
positions = random_walk(N, h)
print(positions[-1])
plt.plot(positions)

positions = random_walk(N, h)
print(positions[-1])
plt.plot(positions)

positions = random_walk(N, h)
print(positions[-1])
plt.plot(positions)

plt.xlabel('Time (steps)')
plt.ylabel('Position')
plt.show()

```

## Code for generating "random_walk.csv"
Run this 

```python
import numpy as np
#np.random.seed(1234)
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 

# Find RMS position for different steps
N = 50 # number of total steps to simulate
step_list = np.floor(np.linspace(1, 10000, N)) #total steps
position_rms = np.zeros(N) # list to store final rms positions

for j in range(N):
    num_simulations = 5000 # number of simulations to run of a given step number
    
    final_positions = np.zeros(num_simulations)
    for i in range(num_simulations):
        positions = random_walk( int(step_list[j]) , h=1)
        final_positions[i] = positions[-1]

    # RMS  
    position_rms[j] = np.sqrt(np.mean(final_positions**2))     
```

```python
# Export data to csv
import pandas as pd
name_dict = {
            'steps':step_list,
            'pos_rms':position_rms
            }
df = pd.DataFrame(name_dict)
df.to_csv('random_walk.csv', index=False)
```

```python
# Plot rms position vs steps
plt.plot(step_list, position_rms)
plt.plot(step_list, np.sqrt(step_list))
plt.xlabel('Steps')
plt.ylabel('Position (RMS)')
plt.show()

# Plot rms position vs steps
plt.plot(step_list, position_rms**2)
plt.plot(step_list, step_list)
plt.xlabel('Steps')
plt.ylabel('Variance')
plt.show()
```

```python
def random_walk(N):

    # make a list of N random values of 1,-1
    random_steps = np.random.choice([1,-1], N) # discrete steps
    
    # add the random steps one-by-one to get a random walk
    positions = np.cumsum(random_steps)  # absolute position
    positions = np.insert(positions, 0, 0.0) # insert the initial position
    
    # Print the final position of the random walk
    print( positions[-1] ) # print the final position
    
    # Return the entire random walk list
    return positions

# Call the function
N=100
steps = random_walk(N) # prints the final position and sets steps equal to the list of all positions
print(steps)
```
